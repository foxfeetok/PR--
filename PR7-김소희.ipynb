**Count Plot**  
*total_bill과 tips과의 관계를 보여주는 scatter plot 그리기*

```python
tit = sns.load_dataset("titanic")
tit.head(5)
```

```python
sns.countplot(x="sex", data=tit)
```

```python
sns.barplot(x="sex", y="survived", data=tit) #살아남은 사람들. 성별로 나누기
```

```python
sns.barplot(x="class", y="survived", data=tit) #클래스 별로 살아남은 사람들 나누기
```

**연속된 변수. Histogram**

```python
#plt.figure(figsize=(8,3)) #그림 사이즈 조정하기 

#나이에 대한 분포도. 
#bin의 갯수로 막대의 간격 조절이 가능 
#kde = 그래프 위 선을 추가
#stat = probability -> 확률
#dodge -> 떨어뜨려 져서 그래프를 그린다. dodge를 쓰지 않는 경우, 누적으로 그래프를 보여준다.

sns.histplot(x="age", bins=30, kde=True, hue="class", multiple="dodge", data=tit) 
```

```python
#class 별로 살아남은 비중에 대해 막대그래프로 그리기

sns.barplot(x="class",y="survived",hue="sex", data=tit)
```

**BOX PLOT**

```python
sns.boxplot(x='class', y='age', hue='sex', data=tit)
```

**Hitmap**

```python
#변수들 간의 상관계수를 구하고 히트맵으로 구하기
#annot -> annotation 으로 숫자를 보여주는 것
#fmt -> format

plt.figure(figsize=(8,8))

corr=tit.corr()
sns.heatmap(corr, annot=True, fmt=".1f")
corr
```

**Tip 에 관한 Data**

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
#요일별 매출 현황
#hue -> smoker 로 구분하여 분류
#palette -> 색을 바꾸어준다

plt.figure(figsize=(9,5))

sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
```

```python
sns.pairplot(hue="smoker", data=tips)
```

#### heatmap
- 참조: https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
flights = sns.load_dataset("flights")
print(flights.head(5))

flights = flights.pivot("month", "year", "passengers")
print(flights.head(5))

plt.figure(figsize=(10, 10))
ax = sns.heatmap(flights, annot=True, fmt="d")
```
